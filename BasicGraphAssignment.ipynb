# Opioids in Virginia - Graphing Questions

### I chose to go with a scatterplot for these graphs because a line or bar graph would become too cumbersome and hard to read. Since we are dealing with multiple categories, I felt these type of graphs would be best fitted to display the data.

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
""" Importing in the Opioids file """

dfopioid = pd.read_csv('C:\\Users\\tfatale\\CTBA\\CTBA\\M2\\OpidsVA.csv')
print('Column Labels: ',dfopioid.columns.values)
```

```python
dfopioid
```

#### Question 1 graph and code

```python
""" Format of the scatterplot method is as follows: ax.scatter(x-series, y-series) """
fig, ax = plt.subplots()
""" 
ax.scatter() statement creates scatterplot
The 'alpha' parameter controls dot transparency: 1 = solid, <1 = various transparency levels, 0 = no mark
The c parameter designates color of the dots: 'b' stands for blue  
"""
ax.scatter(dfopioid['FPOO-Rate'], dfopioid['MedianHouseholdIncome'], alpha=0.5, c = 'b')  
fig.suptitle('Opioid Overdoses vs. Household Income')   # Graph title
ax.xaxis.set_label_text('Opioid Overdoses')  # x-axis caption
ax.yaxis.set_label_text('Total Median Income')       # y-axis caption
fig.set_size_inches(7,5)
plt.show()
```

##### From this data there does not seem to be a direct correlation between the amount of opioid overdoses and household income.

#### Question 2 graph and code

```python
""" Format of the scatterplot method is as follows: ax.scatter(x-series, y-series) """
fig, ax = plt.subplots()
""" 
ax.scatter() statement creates scatterplot
The 'alpha' parameter controls dot transparency: 1 = solid, <1 = various transparency levels, 0 = no mark
The c parameter designates color of the dots: 'b' stands for blue  
"""
ax.scatter(dfopioid['FPOO-Rate'], dfopioid['FFHO-Rate'], alpha=0.5, c = 'b')  
fig.suptitle('Opioid Overdoses vs. Fentanyl/Herione Overdoses')   # Graph title
ax.xaxis.set_label_text('Opioid Overdoses')  # x-axis caption
ax.yaxis.set_label_text('Fentanyl/Herione OVerdoses')       # y-axis caption
fig.set_size_inches(7,5)
plt.show()
```

##### According to the data in this graph there is no direct correlation between the two types of overdoses.

# Presidents Birthplace Question

### I decided that a histogram would be the best way to represent the data. I think that it will display the data better than a scatter plot or boxplot.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
""" Importing in the Presidents Birthplace file """

dfpres = pd.read_csv('C:\\Users\\tfatale\\CTBA\\CTBA\\M2\\presidents.csv')
print('Column Labels: ',dfpres.columns.values)
```

```python
dfpres
```

```python
""" Using pandas .value_counts() method and list comprehension to create th
    frequency histogram data """
births = dict(dfpres['State'].value_counts())
```

```python
births
```

```python
""" Plot frequency histogram data """
fig, ax = plt.subplots()
ax.bar(births.keys(),births.values(), color='g', alpha=0.75)

fig.suptitle('Presidents BirthplaceFrequency Histogram')
fig.set_size_inches(7,5)

ax.xaxis.set_label_text('State')
ax.yaxis.set_label_text('Frequency of Births')
```

##### The birthplace data can be broken into three separate groups. The second and third groups consist of states with the same amount of births (2 births and 1 birth respectively). The first group consists of states that have different amounts of births. These states have the largest amount of Presidents births. They are (in order): Virginia, Ohio, New York, Massachusetts.
